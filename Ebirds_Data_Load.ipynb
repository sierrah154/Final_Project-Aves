```python
# Create list of bird files
csv_file_list = ["hummingbirds.csv", "annas_hummingbird.csv"]
```

```python
# Create path for CSVs
data_folder = Path("./Resources/")
```

```python
# Create dataframes from CSV list and merge into dataframe using concat.
list_of_dataframes = []
for filename in csv_file_list:
    list_of_dataframes.append(pd.read_csv(data_folder/(filename), sep='\t'))

hb_df = pd.concat(list_of_dataframes)

print(hb_df)
```

```python
# View column names
for col in hb_df.columns: 
    print(col) 
```

```python
# Clean columns in dataframe by dropping.

hb_cleaned_df = hb_df.drop(['gbifID', 'datasetKey', 'kingdom', 'phylum', 'class', 'order', \
            'family', 'infraspecificEpithet', 'taxonRank', 'taxonRank', 'scientificName', \
           'verbatimScientificName', 'verbatimScientificNameAuthorship', 'publishingOrgKey', \
           'coordinateUncertaintyInMeters', 'coordinatePrecision', 'elevation', 'elevationAccuracy', \
           'depth', 'depthAccuracy', 'taxonKey', 'speciesKey', 'institutionCode', 'collectionCode', \
           'dateIdentified', 'license', 'rightsHolder', 'recordedBy', 'typeStatus', 'establishmentMeans', \
           'lastInterpreted', 'mediaType', 'issue'], axis=1)
```

```python
hb_cleaned_df.head()
```

```python
# Check data types
hb_cleaned_df.dtypes
```

```python
# Splitting the datetime column
hb_date_df = hb_cleaned_df["eventDate"].str.split("T", n = 1, expand = True) 
  
# New column for date 
hb_cleaned_df["date"]= hb_date_df[0] 
  
# New column for time 
hb_cleaned_df["time"]= hb_date_df[1] 
  
# Dropping old eventDate columns 
hb_cleaned_df.drop(columns =["eventDate"], inplace = True) 
  
# Display DF head 
hb_cleaned_df.head()
```

```python
# Rename columns in dataframe

hb_cleaned_df.rename(columns={
    'occurenceID': 'occurence_id', 
    'genus': 'genus', 
    'species': 'species', 
    'countryCode': 'country_code',
    'locality': 'locality',
    'stateProvince': 'state',
    'occurrenceStatus': 'occurence_status',
    'individualCount': 'individual_count',
    'decimalLatitude': 'latitude',
    'decimalLongitude': 'longitude',
    'day': 'day',
    'month': 'month',
    'year': 'year',
    'basisOfRecord': 'basis_of_record',
    'catalogNumber': 'catalog_number',
    'recordNumber': 'record_number',
    'identifiedBy': 'identified_by',
    'date': 'date',
    'time': 'time'}, inplace=True)
```

```python
hb_cleaned_df.head()
```

```python
# Create DataFrame of locations
hb_locations_df = hb_cleaned_df.filter(['latitude','longitude'], axis=1)
```

```python
# Check length of locations
len(hb_locations_df)
```

```python
# Drop duplicate locations
hb_locations_df.drop_duplicates()
```

```python
# Further clean ebird dataframe by dropping columns
hb_final_df = hb_cleaned_df.drop(['occurrenceID', 'day', 'month', 'year', 'record_number', 'identified_by', 'time'], axis=1)
```

```python
# Further cleaning by dropping na locations
hb_final_df = hb_final_df.dropna(subset=['latitude', 'longitude', 'date'])
```

```python
# Create an initial URL to pass location data through US Census api on FCC to get county.
url = "https://geo.fcc.gov/api/census/block/find?"
```

```python
# Test Api call with first location

location_url = url + "latitude= 33.8283" + "&longitude=-118.343" + "&format=json"
county = requests.get(location_url)
county.json()
```

```python
# Create an empty list to hold the county data.
counties = []
# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1
# Loop through all the locations
i = 0
counties_df = pd.DataFrame(columns = ['latitude','longitude','counties'])
for index,row in hb_locations_df.iterrows():
    location_url = url + "latitude=" + str(row["latitude"]) + "&longitude=" + str(row["longitude"]) + "&format=json"
    response = requests.get(location_url).json()
    county = response["County"]["name"]
    counties.append(county)
 
```

```python
# Create an empty list to hold the county data.
counties = []
# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1
```

```python
# Loop through all the locations
i = 0
for i, row in hb_locations_df.iterrows():
    # Group locations in sets of 1000
    if (i % 1000 == 0 and i >= 1000):
        set_count += 1
        record_count = 1
    # Create URL for each location    
    location_url = url + "latitude=" + str(row["latitude"]) + "&longitude=" + str(row["longitude"]) + "&format=json"
    response = requests.get(location_url).json()
    
    # Pull county name from API
    county = response["County"]["name"]
    
    # Append county to counties list
    counties.append(county)
    
    # Log the URL, record, and set numbers.
    print(f"Processing Record {record_count} of Set {set_count}")
    # Add 1 to the record count.
    record_count += 1
 
```

```python
# Test the first 100 locations
df = hb_locations_df.iloc[0 : 100]
df['county_name']= counties
df.head()
```
